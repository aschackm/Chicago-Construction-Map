```python
import pandas as pd
import numpy as np
```

```python
demo_2010 = pd.read_csv("2010_demographics.csv")
demo_2011 = pd.read_csv("2011_demographics.csv")
demo_2012 = pd.read_csv("2012_demographics.csv")
demo_2013 = pd.read_csv("2013_demographics.csv")
demo_2014 = pd.read_csv("2014_demographics.csv")
demo_2015 = pd.read_csv("2010_demographics.csv")
demo_2016 = pd.read_csv("2010_demographics.csv")
```

```python
demo_2010['Year'] = '2010'
demo_2011['Year'] = '2011'
demo_2012['Year'] = '2012'
demo_2013['Year'] = '2013'
demo_2014['Year'] = '2014'
demo_2015['Year'] = '2015'
demo_2016['Year'] = '2016'
list_of_dfs = [demo_2010, demo_2011, demo_2012, demo_2013, demo_2014, demo_2015, demo_2016]
```

```python
for df in list_of_dfs:
    df.dropna(inplace=True)
    df.rename(columns={'Geography':'Census Tract',
                       'Percent; RACE - One race':'% One Race',
                       'Percent; RACE - One race - White':'% White',
                       'Percent; RACE - One race - Black or African American': "% Black",
                       'Percent; RACE - One race - American Indian and Alaska Native': "% Native American",
                       'Percent; RACE - One race - Asian': "% Asian",
                       'Percent; RACE - One race - Native Hawaiian and Other Pacific Islander': "% Hawaiin/Pacific Islander",
                       'Percent; RACE - One race - Some other race': '% Other',
                       'Percent; RACE - Two or more races': '% Two or More Races'},
              inplace=True)
    for index, row in df.iterrows():
        census_tract_fullstring = row[0]
        census_tract_comma_split = census_tract_fullstring.split(",",maxsplit=1)
        census_tract_string = census_tract_comma_split[0].split(" ", maxsplit=3)
        census_tract_int = float(census_tract_string[2])
        df.loc[index,'Census Tract'] = census_tract_int
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///chidata.sqlite', echo=False)
```

```python
for df in list_of_dfs:
    df_year = df.loc[0,'Year']
    print(df_year)
    df.to_sql(name=f"demo_{df_year}",index=False,con=engine,if_exists='append')
```
